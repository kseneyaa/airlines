```python
data2=data
data2=pd.DataFrame(data2)
```

```python
data2['BirthDate'] = data2['BirthDate'].astype('str')
```

```python
def myDateConv(tt):
    sep = tt[2]
    if len(tt) == 19:
      return pd.to_datetime(tt, format='%Y-%m-%d %H:%M:%S')
    elif sep == '-':
        return pd.to_datetime(tt, format='%Y-%m-%d')
    elif sep == '/':
        return pd.to_datetime(tt, format='%m/%d/%Y')
    else:
        return tt
```

```python
data2['BirthDate'] = data2['BirthDate'].apply(myDateConv)
```

```python
fly_count2 = data2.groupby(['Last__Name', 'First__Name', 'BirthDate'])['PassengerDocument', 'TravelDoc'].agg(['count'])
#fly_count2 ['kol_pasport'] = fly_count2['PassengerDocument count'] + fly_count2['TravelDoc count']
fly_count2
```

```python
fly_count2 ['kol_pasport'] = fly_count2[('PassengerDocument', 'count')] + fly_count2[('TravelDoc', 'count')]
fly_count2
```

```python
fly_count2[(fly_count2['kol_pasport'] > 4)]
```
